```python
def multiply_loops(A, B):
    C = np.zeros((A.shape[0], B.shape[1]))
    for i in range(A.shape[1]):
        for j in range(B.shape[0]):
            C[i, j] = A[i, j] * B[j, i]
    return C
```

```python
def multiply_vector(A, B):
    		return A @ B
```

```python
Y = np.random.random((100, 100))
```

```python
# find sum of these numbers "1,2,3,4,5"
def find_sum(s):
    l_str = s.split(",")
    total = 0
    for e in l_str:
        total += int(e)
    return total

# 
# 15
```
