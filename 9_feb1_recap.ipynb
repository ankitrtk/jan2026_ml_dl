```python
def gcd(a, b):
    r = b % a
    if r == 0:
        return a
    return gcd(r, a)
```

```python
def multiply_eqn(eqn1, mult):
    n_eqn = []
    for i in eqn1:
        n_eqn.append(i*mult)
    return n_eqn
multiply_eqn([1,1,10], 5)
```

```python

def solve1(eqns):
    eqn = eqns[0] # eqn = [3, 4]
    a, b = eqn # a = 3; b = 4
    return [b / a]

eqns = [[3, 4]] # 3x = 4
solve1(eqns)
```

```python
# returns eqn2 - eqn1
def subtract_eqn(eqn2, eqn1):
    result = []
    for i, x in enumerate(eqn2):
        result.append(x - eqn1[i])
    return result
```

```python
def elementwise_mult(eqn1, eqn2):
    # eqn1 = [1, 2, 3]
    # eqn2 = [3, 4, 5]
    # result = [1*3, 2*4, 3*5]
    result = []
    for i, x in enumerate(eqn1):
        result.append(x * eqn2[i])
    return result
```

```python
r = elementwise_mult([3, 2, 7], [5, -5, 8])
print(r)
sum(r)
```

```python
# x + 3*y + 2*z = 20
# y, z = [5, -5]
# x = 20 - (3*5 + 2*-5)

# 4x + 3*y + 2*z + 7u = 20
# y, z, u = [5, -5, 8]
# 4x = 20 - (3*5 + 2*-5 + 7*8)
# x = (20 - (3*5 + 2*-5 + 7*8))/4
def solve_x(eqn1, n_1_vars):
    eqn11 = eqn1[1:-1]
    rhs = eqn1[-1] - sum(elementwise_mult(eqn11, n_1_vars))
    return rhs / eqn11[0]
```

```python
solve_x([4,3,2,7, 20], [5, -5, 8])
```

```python


# x + y = 10
# 2x + y = 15
eqns = [[1, 1, 10], [2, 1, 15]]

def solve2(eqns): 
    eqn1 = eqns[0]
    eqn2 = eqns[1]
    mult = eqn2[0]/eqn1[0]
    eqn11 = multiply_eqn(eqn1, mult)
    eqn21 = subtract_eqn(eqn2, eqn11)[1:]
    n_1_vars = solve1([eqn21])
    x = solve_x(eqn1, n_1_vars) # y = 1, [1, 1, 10], x = 10-1*
    return [x] + n_1_vars
```

```python
eqns = [
    [1, 1, 10],
    [2, 1, 15]
]
solve2(eqns)
```

```python
# x + y = 10
# 2x + y = 15
eqns = [[1, 1, 10], [2, 1, 15]]

def solve3(eqns): 
    if len(eqn) == 1:
        return eqn[-1]/eqn[0]
    eqn1 = eqns[0]
    # eliminate x by subtracting first eqn from second and third after mult coefficients
    new_eqns = []
    for eqn2 in eqns[1:]:
        mult = eqn2[0]/eqn1[0]
        eqn11 = multiply_eqn(eqn1, mult)
        neweqn = subtract_eqn(eqn2, eqn11)[1:]
        new_eqns.append(neweqn)
    # we will be left 2 equn and 2 var -> solve2()
    n_1_vars = solve2([new_eqns])
    x = solve_x(eqn1, n_1_vars) # y = 1, [1, 1, 10], x = 10-1*
    return [x] + n_1_vars
```
