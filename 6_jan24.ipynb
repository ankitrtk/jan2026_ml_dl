```python
def print_table(num):
    print(f'Table [{num}]')
    for i in range(10):
        print(num * (i + 1))
    print("-"*10)
```

```python
print_table(3)
```

```python
def print_all_tables(start, end):
    for num in range(start, end+1):
        print_table(num)
    
```

```python
print_all_tables(21,25)
```

```python
# Find first divisor
def find_first_divisor(num):
    # 49 -> 7
    # 25 -> 5
    # 6 -> 2
    # 63 -> 7
    # 35 -> 5
    i = 2
    mysqrt = int(math.sqrt(num))
    for i in range(2, mysqrt + 1):
        # print(i)
        if num % i == 0:
            return i
    return num
    
```
