```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
kite = np.array([[5, 6],
        [-10, 10],
        [-10, -11],
        [7, -6]])

kite @ np.array([[0, -1],
        [1, 0],])

def plot_kite(kite):
    # Close the shape by repeating the first point
    kite_closed = np.vstack([kite, kite[0]])

    plt.figure()
    plt.plot(kite_closed[:, 0], kite_closed[:, 1], marker='o')
    plt.fill(kite_closed[:, 0], kite_closed[:, 1], alpha=0.2)

    plt.axhline(0)
    plt.axvline(0)
    plt.grid(True)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title("Kite Shape")
    plt.xlabel("X")
    plt.ylabel("Y")

    plt.show()
```

```python
plot_kite(kite)
```

```python
kite1 = kite @ np.array([[0, -1],
        [1, 0],])
```

```python
plot_kite(kite1)
```

```python
kite1 = kite @ np.array([[1, 2],
        [3, 4],])
plot_kite(kite1)
```
