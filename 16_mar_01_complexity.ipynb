```python
# Additions

#We have two number n1 and n2 given as strings. Do the additions by performing digit wise additions.

def to_digit(ds):
    return ord(ds) - ord('0')

def add_digits(n1, n2):
    i, j = len(n1) - 1, len(n2) - 1
    result = []
    carryover = 0
    while i > -1 and j > -1:
        total = to_digit(n1[i]) + to_digit(n2[j]) + carryover
        result.append(str(total%10))
        carryover = total // 10
        i -= 1
        j -= 1
    remaining = n1
    ii = i
    if j > -1:
        remaining = n2
        ii = j
    while ii > -1:
        total = to_digit(remaining[j]) + carryover
        result.append(str(total%10))
        carryover = total // 10
        ii -= 1
    if carryover > 0:
        result.append(str(carryover))
    result.reverse()
    print(''.join(result))
```

```python
add_digits("1234", "5678")
```

```python
# What is the complexity of this HCF?
def hcf(m, n):
    remainder = n % m
    if remainder == 0:
        return m
    else:
        return hcf(remainder, m)
    
```

```python
# What is complexity of this recursive version of factorial?
def fact(n):
    if n == 0:
        return 1
    return n * fact(n - 1)

# Ans: O(n)
```

```python
# Memory complexity

# Memory/Space Complexity: O(1)
# Time complexity: O(n)
def sum_array(arr):
    total = 0
    for x in arr:
        total += x
    return total

# Memory/Space Complexity: O(n)
# Time complexity: O(n)
def create_list(n):
    result = []
    for i in range(n):
        result.append(i)
    return result

# Space: O(1)
# Time: O(n^2)
def print_pairs(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n):
            print(arr[i], arr[j])

# Space: O(n)
# Time: O(n)
def factorial_recur(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

# Space: O(1)
# Time: O(n)
def factorial(n):
    result = 1
    for i in range(2, n+1):
        result *= i
    return result
```
