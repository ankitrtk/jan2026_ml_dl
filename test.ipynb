```python
while True:
    n = (l + r)/2
    print(f"Trying {n} midway between {l} and {r}")
    result = 12 ** n
    print(f"result: {result}")
    if abs(result - target) < 0.0001:
        break
    if result > target:
        r = n
    else:
        l = n
```
