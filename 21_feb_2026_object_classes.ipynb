```python
import numpy as np

import math


def rotate(points, angle):
    theta = angle*math.pi/180
    s, c = math.sin(theta), math.cos(theta)


    # x1 = x cos - y * sin
    # y1 = x sin + y cos

    rot_mat = np.array([
        [c, s],
        [-s, c]
    ])

    return np.matmul(points, rot_mat)
```

```python
import numpy as np
import matplotlib.pyplot as plt

points = np.array([[1, 1], [3, 4]])

plt.ion()  # interactive mode
fig, ax = plt.subplots()

for step in range(36):   # 36 steps = full rotation if 10° each
    ax.clear()

    angle = (step + 1) * 10
    rotated = rotate(points, angle)

    ax.scatter(points[:,0], points[:,1], label="Original")
    ax.scatter(rotated[:,0], rotated[:,1], label=f"Rotated {angle}°")

    # draw lines from origin to points (helps visualize rotation)
    for p in rotated:
        ax.plot([0, p[0]], [0, p[1]])

    ax.set_xlim(-5,5)
    ax.set_ylim(-5,5)
    ax.axhline(0)
    ax.axvline(0)
    ax.set_aspect('equal')
    ax.grid(True)
    ax.legend()

    plt.pause(0.1)

plt.ioff()
plt.show()
```
