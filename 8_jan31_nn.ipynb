```python
X = [[1], [2], [3]] # why [1,2,3] [yrs, domain, education_quality]
y = [1000, 2000, 4500] # salary

def find_closest(x_exp, X, y):
    minimum = None
    closest = None
    for i in range(len(X)):# range(0, 3) -> [0, 1, 2]
        xx = X[i][0]
        diff = abs(xx - x_exp)
        if minimum == None or minimum > diff:
            minimum = diff
            closest = i
    return y[closest]
```

```python
X = [[1, 2], [2, 4], [3, 10]] # why [1,2,3] [yrs, education]
y = [1000, 4000, 45000] # salary

def sq_distance(o1, o2):
    x1, y1 = o1
    x2, y2 = o2
    return (y2-y1)**2 + (x2 - x1)**2 # abs(y2-y1) + abs(x2-x1) why not this??

def find_closest_2d(x_new, X, y, distance=sq_distance):
    minimum = None
    closest = None
    for i, xx in enumerate(X):
        dist = distance(x_new, xx)
        if minimum == None or dist < minimum:
            minimum = dist
            closest = i
    return y[closest]
```

```python
print(find_closest_2d([2, 10], X, y)) # to which param we are going to give more weight??

def abs_dist(o1, o2):
    x1, y1 = o1
    x2, y2 = o2
    return abs(y2-y1) + abs(x2-x1)

print(find_closest_2d([2, 10], X, y, abs_dist))
```
