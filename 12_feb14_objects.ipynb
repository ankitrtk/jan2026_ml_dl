```python
class LineModel: # y = mx + c
    def fit(self, data):
        x1, y1 = data[0]
        x2, y2 = data[1]
        # y2 = m * x2 + c
        # y1 = m * x1 + c

        self.m = (y2 - y1)/(x2 - x1)
        self.c = y2 - self.m * x2

    def predict(self, x_new):
        # return x_new*m + c
        return self.m*x_new + self.c
```

```python
# print warning after 5 times predict , "Model might be stale"
class LineModel: # y = mx + c
    predictions_since_training = 0
    def fit(self, X, y):
        x1, y1 = X[0][0], y[0]
        x2, y2 = X[1][0], y[1]
        # y2 = m * x2 + c
        # y1 = m * x1 + c

        self.m = (y2 - y1)/(x2 - x1)
        self.c = y2 - self.m * x2
        self.predictions_since_training = 0

    def predict(self, X_new):
        result = []
        for row in X_new:
            self.predictions_since_training += 1
            if self.predictions_since_training > 5:
                print(f"Warning! The model might be stale. Count: {self.predictions_since_training}")
            x = row[0]
            y = self.m*x + self.c
            result.append(y)
        return result
```

```python
model = LineModel()
X = [
    [2], 
    [4]]
y = [
    3, 
    4]
model.fit(X, y)

X_new = [[2.5], [4]]
print(model.predict(X_new)) # it should print 3
X_new = [[2.5], [4]]
print(model.predict(X_new)) # it should print 3
X_new = [[2.5],]
print(model.predict(X_new)) # it should print 3
X_new = [[2.5], [4], [9]]
print(model.predict(X_new)) # it should print 3
X_new = [[2.5], [4], [9]]
print(model.predict(X_new)) # it should print 3
```
